Task 3: Decision Tree Classification (Example - Iris or Wine)

```python
# Task 3: Decision Tree Classification (Example - Iris or Wine)
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris  # or load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Experiment with parameters
params = [
    {"max_depth": None, "min_samples_split": 2, "criterion": "gini"},
    {"max_depth": 3, "min_samples_split": 5, "criterion": "entropy"}
    # add more combinations
]

for param in params:
    dt = DecisionTreeClassifier(**param, random_state=42)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    print(f"\nParams: {param}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

    # Visualize tree (very important for analysis!)
    plt.figure(figsize=(12,8))
    plot_tree(dt, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
    plt.title(f"Decision Tree - max_depth={param['max_depth']}")
    plt.show()
```
